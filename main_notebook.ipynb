```python
import pandas as pd
import numpy as np

# Imported the necessary libraries
```

```python
df = pd.read_csv(r'Copy of My year - Log.csv', header=None)
df

# Reading Garr's CSV document
```

## Now that the CSV is loaded into a dataframe, lets try to wrangle the data

```python
df.drop([0])
```

```python
averageRating = pd.DataFrame(df, columns=[0, 27])
averageRating
```

```python
averageRating = averageRating.drop([0,1,2])
averageRating
```

```python
averageRating = averageRating.rename(columns={0: 'Day', 27: 'Rating'})
averageRating
```

```python
import matplotlib.pyplot as plt
```

```python
averageRating.plot(x ='Day', y='Rating', kind = 'line')
plt.show()
```

```python
averageRating.dtypes
```

```python
averageRating['Rating']=averageRating['Rating'].astype(float)
averageRating
```

```python
averageRating.plot(x ='Day', y='Rating', kind = 'scatter')
plt.show()
```

```python
x = averageRating['Day']
y= averageRating['Rating']
plt.Line2D(x, y)
plt.xlabel('Day')
plt.ylabel('The Days Rating')
plt.show()  
```
